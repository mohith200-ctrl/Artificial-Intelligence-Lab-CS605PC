```python
def hello(state):
        print("Initial State is: " ,state)
        state["monkey"] = state["box"]
        state["box"] = state["banana"]
        state["banana"] = state["box"]
        print("Goal State is:" ,state)

one = input("enter monkey position")
two = input("enter box position")
three = input("enter banana position")
state = {'monkey':one, 'box':two, 'banana':three}
hello(state)


```
